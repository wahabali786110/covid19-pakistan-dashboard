```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
covid=pd.read_csv('owid-covid-data.csv')
```

```python
covid=covid[['location','date','total_cases','new_cases','total_deaths','new_deaths','new_cases_smoothed','new_deaths_smoothed','people_vaccinated','people_fully_vaccinated']]
```

```python
pakistan=covid[covid['location']=='Pakistan']
```

```python
pakistan['date']=pd.to_datetime(pakistan['date'])
```

```python
pakistan['Year']=pakistan['date'].dt.year
pakistan['Month']=pakistan['date'].dt.month_name()
pakistan['Day']=pakistan['date'].dt.day_name()
```

```python
pakistan.groupby('Year')['new_deaths'].sum()
```

```python
pakistan['total_cases']=pakistan['total_cases'].ffill()
```

```python
pakistan=pakistan[pakistan['total_cases']>0]
```

```python
pakistan['total_deaths']=pakistan['total_deaths'].ffill()
```

```python
pakistan['people_vaccinated']=pakistan['people_vaccinated'].fillna(0)
pakistan['people_fully_vaccinated']=pakistan['people_fully_vaccinated'].fillna(0)
```

```python
pakistan['total_deaths']=pakistan['total_deaths'].fillna(0)
```

# Plot testing

```python
fig=px.line(
        pakistan,
        x='date',
        y='new_cases',
        title='COVID-19 Infection Waves in Pakistan',
        labels={'date':'Date','new_cases':'Daily New Cases'},
        template='plotly_dark'
            )

fig.show()
```
